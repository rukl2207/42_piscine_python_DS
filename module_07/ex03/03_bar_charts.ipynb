```python
query = "select timestamp from checker where uid like 'user_%'"
df = pd.io.sql.read_sql(query, connection, parse_dates='timestamp')
connection.close()
df
```

```python
df['date'] = df['timestamp'].dt.date
df['hour'] = df['timestamp'].dt.hour
df['dayofweek'] = df['timestamp'].dt.dayofweek
df['weekday'] = df['dayofweek'].apply(lambda x: 'working_day' if x < 5 else 'weekend')
df
```

```python
df1 = df.groupby(['weekday', 'date', 'hour']).count()
df1 = df1.groupby(['weekday', 'hour']).mean().reset_index()
df1
```

```python
df2 = pd.pivot(df1, index='hour', columns='weekday', values='dayofweek')
df2
```

```python
ax = df2.plot(kind='bar', subplots=True, fontsize=8, figsize=(15,8), title="Commits per day", colormap='Set3', rot=0)
```
