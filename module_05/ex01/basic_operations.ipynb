# Exercise 01 : Basic operations

```python
import pandas as pd
```

## create a dataframe views with two columns: datetime and user by reading feed-views.log

```python
df = pd.read_csv("../data/feed-views.log",
                 names=("datetime", "user"),
                 sep="\t",
                 engine="python")
df.head()
```

```python
df.info()
```

```python
df["datetime"] = pd.to_datetime(df["datetime"])
df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df["hour"] = df["datetime"].dt.hour
df["minute"] = df["datetime"].dt.minute
df["second"] = df["datetime"].dt.second
df.head()
```

```python
df.info()
```

## create the new column daytime

```python
labels = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
bins = [0, 4, 7, 11, 17, 20, 24]
df["daytime"] = pd.cut(df["hour"], labels=labels, bins=bins, include_lowest=True, right=False)
df.set_index("user", inplace=True)
df.head(50)
```

```python
df.info()
```

## calculate the number of elements in your dataframe

```python
df.count()
```

```python
df.value_counts("daytime")
```

## sort values in your dataframe by hour, minute, and second in ascending order

```python
df.sort_values(by=["hour", "minute", "second"])
```

## calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
max_night = df[df["daytime"] == "night"]["hour"].max()
min_morning = df[df["daytime"] == "morning"]["hour"].min()
max_night, min_morning
```

```python
df[(df["daytime"] == "night") & (df["hour"] == 1)][:1]
```

```python
df.loc[df.daytime == 'night'].hour.idxmax()
```

```python
df[(df["daytime"] == "morning") & (df["hour"] == 8)][:1]
```

```python
df.loc[df.daytime == 'morning'].hour.idxmin()
```

```python
mode_hour = df.hour.mode()
mode_hour
```

```python
mode_daytime = df.daytime.mode()
mode_daytime
```

## show the 3 earliest hours in the morning and the corresponding usernames and the 3 latest hours and the usernames

```python
earliest = df[df["daytime"] == "morning"].nsmallest(3, ["hour", "minute", "second"])
earliest
```

```python
latest = df.nlargest(3, ["hour", "minute", "second"])
latest
```
